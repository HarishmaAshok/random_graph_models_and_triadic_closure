# triadic closure

##### generating a random graph and finding the no of nodes, links, adjacent nodes and degree

```python
# generating a random graph
import itertools   
nodes = 5
probability = 0.5
G.clear()
G = nx.erdos_renyi_graph(nodes, probability)
nx.draw(G, with_labels = True)
print("\nnumber of nodes in the graph:\n",G.nodes)
print("======================================================================================")

print("\nnumber of edges/links in the graph:\n",G.edges)
print("======================================================================================")

print("\nlinks from every node:")
for node in range(0, nodes):
    print("node {} : {}".format(node, nx.edges(G, [node])))
print("======================================================================================")
   
print("\nList of all nodes we can go to in a single step from")
newEdgeList = []
for val in range(0,nodes):
    print("\nnode {}: {}".format(val, list(G.neighbors(val))))
    aa = list(itertools.combinations(list(G.neighbors(val)), 2))
    newEdgeList.append(aa)
    print("adjacent nodes possible link combinations:",aa)

print("======================================================================================")
from collections import OrderedDict
print("\nDegree for all nodes:\n",OrderedDict(sorted(dict(G.degree()).items())))
```

```python
def triadic_closure():
    a = list(G.edges)

    for i in range(0, len(a)):
        one = a[i]
        for j in range(i+1, len(a)):
            two = a[j]
            print(one, two)
            if(one[0] == two[0]):
    #             print("first index value is same")
                ne = (one[1],two[1])
    #             print(ne, a)
    #             print((ne in a))
                if (((ne in a) is not True) and (ne[::-1] in a) is not True):
                    print("link doesnot exists. creating link with {}".format(ne))
                    G.add_edge(*ne)
                    print(ne)

            if(one[0] == two[1]):
                ne = one[1],two[0]
                if (((ne in a) is not True) and (ne[::-1] in a) is not True):
                    print("link doesnot exists. creating link with {}".format(ne))
                    G.add_edge(*ne)
                    print(ne)

            if(one[1] == two[1]):
                ne = one[0],two[0]
                if (((ne in a) is not True) and (ne[::-1] in a) is not True):
                    print("link doesnot exists. creating link with {}".format(ne))
                    G.add_edge(*ne)
                    print(ne)

            if(one[1] == two[0]):
                ne = one[0],two[1]
                if (((ne in a) is not True) and (ne[::-1] in a) is not True):
                    print("link doesnot exists. creating link with {}".format(ne))
                    G.add_edge(*ne)
                    print(ne)

            print("\n")

    nx.draw(G, with_labels = True)

triadic_closure()
```
